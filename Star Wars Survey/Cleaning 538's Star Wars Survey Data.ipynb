```python
import pandas as pd
star_wars = pd.read_csv("data/StarWars.csv", encoding="ISO-8859-1")
star_wars.head(3)
```

```python
star_wars.shape
```

```python
star_wars.columns
```

Remove respondents with no unique id

```python
star_wars = star_wars[star_wars["RespondentID"].notnull()]
star_wars.shape
```

```python
yes_no = {"Star Wars: Episode I  The Phantom Menace": True,
 "Star Wars: Episode II  Attack of the Clones": True,
 "Star Wars: Episode III  Revenge of the Sith": True,
 "Star Wars: Episode IV  A New Hope": True,
 "Star Wars: Episode V The Empire Strikes Back": True,
 "Star Wars: Episode VI Return of the Jedi": True,
          'NaN' : False}


for i in range(3,9,1):
    star_wars[star_wars.columns[i]] = star_wars[star_wars.columns[i]].map(yes_no)

display(star_wars.head(10))
```

```python
display(star_wars.columns[3:9])
```

```python
col_names = {
    'Which of the following Star Wars films have you seen? Please select all that apply.': "seen_1",
    'Unnamed: 4': "seen_2", 
    'Unnamed: 5': "seen_3", 
    'Unnamed: 6': "seen_4", 
    'Unnamed: 7': "seen_5",
    'Unnamed: 8': "seen_6"
}
star_wars = star_wars.rename(columns=col_names)
star_wars.head()
```

```python
star_wars[star_wars.columns[9:15]] = star_wars[star_wars.columns[9:15]].astype(float)
```

```python
display(star_wars.columns[9:15])
```
