Reading in the Survey Text Files

```python
all_survey = pd.read_csv("data/survey_all.txt", delimiter="\t", encoding="windows-1252")
all_survey.head(2)
survey_d75 = pd.read_csv("data/survey_d75.txt", delimiter="\t", encoding="windows-1252")
survey = pd.concat([all_survey, survey_d75], axis=0)
survey.head(2)
```

```python
survey["DBN"] = survey["dbn"]
```

```python
cols = ["DBN", "rr_s", "rr_t", "rr_p", "N_s", "N_t", "N_p", "saf_p_11", "com_p_11", "eng_p_11", "aca_p_11", "saf_t_11", "com_t_11", "eng_t_11", "aca_t_11", "saf_s_11", "com_s_11", "eng_s_11", "aca_s_11", "saf_tot_11", "com_tot_11", "eng_tot_11", "aca_tot_11"]

survey = survey[cols]

data["survey"] = survey
data["survey"].info()
```

#### The next code standardizes the hs_directory file dbn column to DBN, and then, since, class_size has DBN colum, we identify that since DBN in the sat_results data is just a combination of the CSD and SCHOOL CODE columns in the class_size data - we use this column to generate a DBN column.

```python
data["hs_directory"]["DBN"] = data["hs_directory"]["dbn"]

def pad_csd(num):
    num_str = str(num)
    if len(num_str) > 1:
        return num_str
    else:
        return num_str.zfill(2)
    
data["class_size"]["padded_csd"] = data["class_size"]["CSD"].apply(pad_csd)
data["class_size"]["DBN"] = data['class_size']["padded_csd"] + data["class_size"]['SCHOOL CODE']
data["class_size"].head()
```

#### Below, we sum up 3 columns of SAT results file to produce a single sat_score that we can use to correlate with other factors

```python
cols = ["SAT Math Avg. Score", "SAT Critical Reading Avg. Score", "SAT Writing Avg. Score"]
for col in cols:
   data["sat_results"][col] = pd.to_numeric(data["sat_results"][col], errors="coerce")
    
data["sat_results"]["sat_score"] = data["sat_results"][cols[0]] + data["sat_results"][cols[1]] + data["sat_results"][cols[2]]
data["sat_results"]["sat_score"].head()
```

```python
Extracting Latitude from hs_directory
```

```python
import re

def extract_coordinates(string):
    result_string = re.findall("\(.+\)", string)[0].split(",")[0]
    return result_string.replace("(", "")


data["hs_directory"]["lat"] = data["hs_directory"]["Location 1"].apply(extract_coordinates)

data["hs_directory"].head()
```

```python
import re

def extract_coordinates(string):
    result_string = re.findall("\(.+\)", string)[0].split(",")[1]
    return result_string.replace(")", "")


data["hs_directory"]["lon"] = data["hs_directory"]["Location 1"].apply(extract_coordinates)

data["hs_directory"].head()
```

```python
cols = ["lat", "lon"]

for col in cols:
    data["hs_directory"][col] = pd.to_numeric(data["hs_directory"][col], errors="coerce")
```
